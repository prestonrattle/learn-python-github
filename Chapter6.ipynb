```python
# we can iterate over a file, aggregating the value counts in the "key" column

chunker = pd.read_csv('downloads/ex6.csv', chunksize = 1000)
tot = pd.Series([], dtype = 'int64')

for piece in chunker:
    tot = tot.add(piece['key'].value_counts(), fill_value = 0)

tot = tot.sort_values(ascending = False)
tot
```
