```python
import numpy as np
import pandas as pd
```

```python
# sentinel value = missing value indicator

float_data = pd.Series([1.2, -3.5, np.nan, 0])
float_data
```

```python
# .isna() method checks for missing data

float_data.isna()
```

```python
# built in Python "None" is treated as NA

string_data = pd.Series(['aardvark', np.nan, None, 'avocado'])
string_data
```

```python
string_data.isna()
```

```python
float_data = pd.Series([1, 2, None], dtype = 'float64')
float_data.isna()
```

```python
# .dropna() method returns the Series with only the nonnull and index values

data = pd.Series([1, np.nan, 3.5, np.nan, 7])
data.dropna()
```

```python
# this is the same thing as above

data[data.notna()]
```

```python
data = pd.DataFrame([[1., 6.5, 3.],
                     [1., np.nan, np.nan],
                     [np.nan, np.nan, np.nan],
                     [np.nan, 6.5, 3.]])
data
```

```python
# for DataFrames, .dropna() drops any row containing missing values

data.dropna()
```

```python
# how = 'all' only drops rows where all values are missing

data.dropna(how = 'all')
```

```python
data[4] = np.nan
data
```

```python
# drop columns with the same logic

data.dropna(axis = 'columns')
```

```python
data.dropna(axis = 'columns', how = 'all')
```

```python
df = pd.DataFrame(np.random.standard_normal((7, 3)))
df
```

```python
df.iloc[:4, 1] = np.nan
df.iloc[:2, 2] = np.nan
df
```

```python
df.dropna()
```

```python
# only keep rows with a certain number of missing values, using the "thresh" argument

df.dropna(thresh = 2)
```

```python
# use the .fillna() method to replace missing values instead of dropping them
# can supply a constant value to use

df.fillna(value = 100)
```

```python
# call .fillna() with a dictionary to use a different fill value for each column

df.fillna(value = {0:10, 1:20, 2:30})
```

```python
# interpolation methods can be used

df = pd.DataFrame(np.random.standard_normal((6, 3)))
df.iloc[2:, 1] = np.nan
df.iloc[4:, 2] = np.nan
df
```

```python
# forward fill .. repeat last non-null value

df.ffill()
```

```python
# can set a limit on how many values to forward fill

df.ffill(limit = 2)
```

```python
# can also use summary stats like mean or median

data = pd.Series([1., np.nan, 3.5, np.nan, 7])
data.fillna(value = data.mean())
```

```python
data = pd.DataFrame({"k1": ["one", "two"] * 3 + ["two"],
                     "k2": [1, 1, 2, 3, 3, 4, 4]})
data
```

```python
# .duplicated() method returns a Boolean Series if the row is a duplicate (across columns)

data.duplicated()
```

```python
# drop duplicate values

data.drop_duplicates()
```

```python
# by default, .duplicates() considers all columns
# can adjust this to be a subset of columns

data['v1'] = range(7)
data
```

```python
# drop duplicates based on the k1 column
# keeps just the first observed non-duplicate per the subset

data.drop_duplicates(subset = ['k1'])
```

```python
# "keep" argument will just return the last duplicate

data.drop_duplicates(subset = ['k1'], keep = 'last')
```

```python
data = pd.DataFrame({"food": ["bacon", "pulled pork", "bacon",
                              "pastrami", "corned beef", "bacon",
                              "pastrami", "honey ham", "nova lox"],
                     "ounces": [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
# to create new columns based on a function or mapping, use the .map() method

# first we create the mapping
meat_to_animal = {
  "bacon": "pig",
  "pulled pork": "pig",
  "pastrami": "cow",
  "corned beef": "cow",
  "honey ham": "pig",
  "nova lox": "salmon"
}
```

```python
# .map() method for Series accepts a function or dictionary-like object with mapping

data['animal'] = data['food'].map(meat_to_animal)
data
```

```python
# could also pass a function

def get_animal(x):
    return meat_to_animal[x]

data['food'].map(get_animal)
```

```python
data = pd.Series([1., -999., 2., -999., -1000., 3.])
data
```

```python
# if the -999 are actually missing values, we can replace them

data.replace(-999, np.nan)
```

```python
# can replace multiple values with one value

data.replace([-999, -1000], np.nan)
```

```python
# can map replacements with a dictionary too

data.replace({-999:np.nan, 1:np.nan})
```

```python
data = pd.DataFrame(np.random.standard_normal((1000, 4)))
data.describe()
```

```python
# suppose you wanted to find values exceeding 3 in absolute value in one of the columns

col = data[2]
col[col.abs() > 3]
```

```python
# select all rows having a value > ABS 3, use the .any() method
# parentheses around data.abs() > 3 are necessary in order to call the any method on the result of the comparison operation

data[(data.abs() > 3).any(axis = 'columns')]
```

```python
# cap values outside the interval -3 to 3

data[data.abs() > 3] = np.sign(data) * 3
data.describe()
```

```python
# np.sign(data) produces 1 and -1 values based on whether the values are positive or negative

np.sign(data).head()
```

```python
df = pd.DataFrame(np.arange(5 * 7).reshape((5, 7)))
df
```

```python
# permuting = random reordering
# numpy.random.permutation() function produces an array of integers with a random ordering
# when passed an integer, creates an array that long, otherwise randomly reorders

sampler = np.random.permutation(5)
sampler
```

```python
# can use array with .iloc[] to index

df.iloc[sampler]
```

```python
# same thing as df.iloc[sampler]

df.take(sampler)
```

```python
column_sampler = np.random.permutation(7)
column_sampler
```

```python
# can also select a permutation of the columns

df.take(column_sampler, axis = 'columns')
```

```python
# to select a random subset with replacement use the .sample() method

df.sample(3)
```

```python
# to generate a sample with replacement, use the "replace" argument in the .sample() method

choices = pd.Series([5, 7, -1, 6, 4])
choices.sample(n = 10, replace = True)
```

```python
df = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                   'data1': range(6)})
df
```

```python
# use pandas.get_dummies() function to get dummy variables

pd.get_dummies(df['key'], dtype = 'float')
```

```python
# can add a prefix to the dummy values

dummies = pd.get_dummies(df['key'], dtype = 'float', prefix = 'key')

# then join to a DataFrame
df_with_dummy = df[['data1']].join(dummies)
df_with_dummy
```

```python
mnames = ['movie_id', 'title', 'genres']

movies = pd.read_table('downloads/movies.dat', sep = '::', header = None, names = mnames, engine = 'python')

movies
```

```python
# .str.get_dummies() method can take delimited groups in one string into multiple dummy variables

dummies = movies['genres'].str.get_dummies('|')
dummies.iloc[:10, :6]
```

```python
# combine dummies with the movies DataFrame
# can also add prefixes

movies_windic = movies.join(dummies.add_prefix('Genre_'))
movies_windic.iloc[0]
```

```python
# useful combo is to combine pandas.get_dummies() with a binning function like pandas.cut()

np.random.seed(12345)
values = np.random.uniform(size = 10)
values
```

```python
bins = [0, 0.2, 0.4, 0.6, 0.8, 1]

pd.get_dummies(pd.cut(values, bins), dtype = 'int')
```
